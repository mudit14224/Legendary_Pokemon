```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
dataset = pd.read_csv("pokemon.csv", sep=",")
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
corr = dataset.corr()
corr.style.background_gradient()
```

```python
fea = ["attack", "base_egg_steps", "base_total", "defense", "experience_growth", "height_m", "hp", "sp_attack", "sp_defense", "speed", "weight_kg"]
```

```python
labels = dataset["is_legendary"]
features = dataset[fea]
```

```python
features.isnull().any()
```

```python
print(features["height_m"].mean())
```

```python
print(features["weight_kg"].mean())
```

```python
features['height_m'] = features['height_m'].fillna(features["height_m"].mean()) 
features['weight_kg'] = features['weight_kg'].fillna(features["weight_kg"].mean())
```

```python
features.isnull().any()
```

```python
corr_f = features.corr()
corr_f.style.background_gradient()
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size = 0.25, random_state = 7)
```

```python
tree = DecisionTreeClassifier(random_state = 5)
```

```python
tree.fit(X_train, y_train)
```

```python
pred = tree.predict(X_test)
pred
```

```python
accuracy_score(y_true=y_test, y_pred=pred)
```

```python
print(classification_report(y_test, pred))
```

```python
model_a = RandomForestClassifier()
X_train,X_test,y_train,y_test = train_test_split(features,labels,test_size=0.25, random_state=7)
model_a.fit(X_train,y_train)
model_a.score(X_test,y_test)
```
